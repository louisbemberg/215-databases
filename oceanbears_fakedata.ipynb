# Person (Superclass)

```python
# creates an array of dictionaries, where each dict is a row in DB
# 100 rows by default with id = 1..num_rows

def create_person(num_rows=100):
    persons = []
    for i in range(num_rows):
        person = {
            'person_id': i+1,
            'p_name': fake.name(),
            'p_phone_number': fake.phone_number(),
            'p_address': fake.address()
        }
        persons.append(person)
        
    return persons
```

```python
# creating 20 surf shops
def create_shop(num_rows=20):
    elements = []
    opening_times = ['7:00', '7:30', '8:00', '8:30', '9:00']
    closing_times = ['16:30', '17:00', '17:30', '18:00', '18:30']
    for i in range(num_rows):
        element = {
                'shop_id': i+1,
                's_name': f"{fake.city()} Surf Shop",
                's_address': fake.address(),
                's_phone_number': fake.phone_number(),
                'opening_time': random.sample(opening_times, 1)[0],
                'closing_time': random.sample(closing_times, 1)[0]
            }
        elements.append(element)
        
    return elements
```

#Â Inventory (Superclass)

```python
def create_inventory(num_rows=2000):
    elements = []
    
    
    for i in range(num_rows):
        element = {
            # id from 1 to 2000
            'IID': i+1,
            #Â making sure the store has a bit of everything
            'is_of_type_item': (i%50 + 1),
            # 50 inventory items per shop, 20 shops in total
            'owned_by_shop': (i%20 + 1)
        }
        elements.append(element)
        
    return elements
```

# Inventory (Subclasses)

```python
def create_inventory_for_rent(num_rows=1000):
    elements = []
    
    # first 1000/2000 elements of the DB are for rent
    for i in range(num_rows):
        giver_id = random.randint(1,100)
        
        # making sure a person isn't renting from themselves
        while True:
            taker_id = random.randint(1,100)
            if giver_id != taker_id: break
        
        hourly_price = random.randint(10, 50)
        element = {
            #Â assigns ids from 1 to 1000
            'IID': i+1,
            
            'is_of_type_item': (i%50) + 1,
            'half_day_price': round(hourly_price * 4 * 0.75),
            'day_price': round(hourly_price * 8 * 0.6),
            'hourly_price': hourly_price,
            'belongs_to_person': giver_id,
            'rented_by_person': taker_id,
            # deposit is a multiple of 5
            'amount_deposit': random.randrange(10, 100, 5),
        }
        elements.append(element)
        
    return elements

def create_inventory_for_sell(num_rows=1000):
    elements = []
    
    # Next 1000 elements of the DB are for selling
    for i in range(1001, 1001+num_rows):
        element = {
            # id from 1001 to 2000
            'IID': i,
            #Â randomized item id from 1 to 50
            'is_of_type_item': (i%50),
            'price': random.randrange(50, 350, 5),
            # randomized person because it doesn't matter
            'bought_by_person': random.randint(1,100)
        }
        elements.append(element)
        
    return elements
```
