```python
# loading necessary packages
import numpy as np
import pandas as pd
import datetime
import random
from faker import Faker
```

```python
# generating faker object with english settings
fake=Faker('en_US')
```

# Person (Superclass)

```python
# creates an array of dictionaries, where each dict is a row in DB
# 100 rows by default with id = 1..num_rows

def create_persons(num_rows=100):
    persons = []
    for i in range(num_rows):
        person = {
            'person_id': i+1,
            'p_name': fake.name(),
            'p_phone_number': fake.phone_number(),
            'p_address': fake.address()
        }
        persons.append(person)
        
    return persons
```

```python
# Testing it out
pd.DataFrame(create_persons())
```

# Person Subclasses (surf_teacher, surfer, event_staff, other_person)


```python
# the first 10 persons of the db are surf teachers
def create_surf_teachers(num_rows=10):
    surf_teachers = []
    for i in range(num_rows):
        surf_teacher = {
                # assign id 1..10 to the teachers
                'person_id': i+1,
                'hourly_rate': random.randint(15, 50)
            }
        surf_teachers.append(surf_teacher)
        
    return surf_teachers
```

```python
# the next 40 persons (id 11...50) of the db are surfers
def create_surfers(num_rows=40):
    surfers = []
    for i in range(num_rows):
        surfer = {
                'person_id': 10+(i+1),
            }
        surfers.append(surfer)
        
    return surfers
```

```python
# creating 7 manufacturers
def create_manufacturers(num_rows=7):
    elements = []
    brands = ['Quiksilver', 'Ripcurl', 'Billabong', "O'Neill", 'Volcom', 'Vans', 'Hurley']
    for i, brand in enumerate(brands):
        element = {
                # assign id 1..10 to the teachers
                'manufacturer_id': i+1,
                'm_name': brand
            }
        elements.append(element)
        
    return elements
```
