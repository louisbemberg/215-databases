```python
# function that creates a random datetime (date + time)
def generate_random_datetime(min_year=2020, max_year=dt.now().year):
    # generate a datetime in format yyyy-mm-dd hh:mm:ss.000000
    start = dt(min_year, 1, 1, 0, 0, 0)
    years = max_year - min_year + 1
    end = start + timedelta(days=365 * years)
    return start + (end - start) * random.random()
```

```python
# 100 sample lessons
def create_lesson(num_rows=100):
    elements = []
    
    for i in range(num_rows):
        while True:
            random_start_datetime = generate_random_datetime()
            # lessons cannot start before 8AM or after 4PM.
            if 8 <= random_start_datetime.hour <= 16: break
        
        # lessons last between 1 and 1 hour (simplified to make life easier)
        # lesson_length = random.randint(1,3)
        lesson_length = 1
        
        # Student level ranges from 1 to 10
        level = random.randint(1, 10)
        
        # Lesson id
        lesson_id = i + 1
        
        # an instructor teaches up to 5 students at the same time
        num_students = random.randint(1,5)
        
        # Teacher id based on lesson id
        if lesson_id % 2 == 0:
            teacher_id = random.randrange(1, 11, 2)
        
        if lesson_id % 2 == 1:
            teacher_id = random.randrange(2, 11, 2)
        
        element = {
            # added to make my life easier hihi
            'lesson_id': lesson_id,
            # using the function above.
            # Note: the dates are shipped in datetime format
            'start_time': random_start_datetime,
            # remember that teachers have PIDs ranging from 1 to 10
            # Note: tiny chance that a teacher may teach 2 lessons at the same time
            'is_given_by_surf_teacher': teacher_id,
            # randomly chooses the id of 1 of the 30 locations of the DB
            'beach_location': random.randint(1, 20), 
            # lessons randomly finish between 1h and 3h after start_time
            'end_time': random_start_datetime + timedelta(hours=lesson_length),
            'max_participants': num_students,
            'student_level': level,
            # for simplicity, every lesson costs 35 dollars
            'total_price_paid': 35 * num_students
                 
        }
        elements.append(element)
        
    return elements
```

# Takes Lesson

```python
# 100 sample lessons
def create_takes_lesson(num_rows=100):
    elements = []
    i = 0
    id_person = {}
    
    while i < num_rows:
        
        # Generate random lesson id
        lesson_id = random.randint(1, 100)
        
        # Create the list of person for the lesson
        if lesson_id not in id_person:
            id_person[lesson_id] = []
            
        # Generate random person id
        person_id = random.randint(1, 100)
        
        if person_id not in id_person[lesson_id]:
            # We add the person to the lesson
            id_person[lesson_id].append(person_id)
            
            # We increment i
            i += 1
            
            if lesson_id % 2 == 0:
                review = round(random.uniform(3, 5), 1)
                
            if lesson_id % 2 == 1:
                review = round(random.uniform(1, 3), 1)
            
            element = {
                # connecting it to one of the lessons
                'lesson_id': lesson_id,
                # 'review': random.sample(['fantastic', 'amazing', 'so good!', 'awful', 'terrible', 'the best ever!'], 1)[0],
                'review' : review,
                # fixed price for 1 hour, can add complexity later on
                'price': 20,
                'person_id': person_id
            }
            
            elements.append(element)
        
    return elements
```

# registers_for_event

```python
# assumes that someone can register for an event altho the event may be full
# otherwise a bit too tricky to fetch the randomized max_capacity
# Note: made the event capacity 300-1000 so it is impossible to overflow an event
def create_registers_for_event(num_rows=250):
    elements = []
    
    event_registrations = {(k+1): [] for k in range(100)}
    
    for i in range(num_rows):
      
        if i < 100:
            event_id = random.randint(1, 100)
            person_id = i+1
            event_registrations[person_id].append(event_id)

        else:
            person_id = random.randint(1,100)
            while True:
                event_id = random.randint(1, 100)
                if event_id not in event_registrations[person_id]:
                    event_registrations[person_id].append(event_id)
                    break
        
        
        element = {
            # connecting it to one of the events
            'event_id': event_id,
            # small risk that the same person may register for the same event twice
            'person_id': person_id,
            # rating can range from 0-5
            'rating': random.randint(1, 5)
        }
        elements.append(element)
    
    return elements
```

# Makes_transaction

```python
# about 500 transactions (there is a total of 2000 inventory items)
def create_makes_transaction(num_rows=500):
    elements = []
    
    # once an item is purchased it cannot be purchased again
    purchased = []
    
    # keeping track of the season for the transactions
    season_transactions = {
        'winter': 0,
        'spring': 0,
        'fall': 0,
        'summer': 0

    }
    
    # we want the sales in summer to > fall == spring > winter
    # for a total of 500 transactions:
    season_standards = {
        'winter': 30,
        'spring': 80,
        'fall': 80,
        'summer': 310
    }
    
    for i in range(num_rows):
        person_id = random.randint(1, 100)
        
        # making sure that the quota of seasons is reached
        while True:
            datetime_out = generate_random_datetime(min_year = 2021)
            month = datetime_out.month
            if month in [12, 1, 2]:
                season = 'winter'
            elif month in [3,4,5]:
                season = 'spring'
            elif month in [6,7,8]:
                season = 'summer'
            elif month in [9, 10, 11]:
                season = 'fall'
            
            if season_transactions[season] < season_standards[season]:
                season_transactions[season] += 1
                break
                
        
        while True:
            # pick an inventory item at random
            inventory_item_id = random.randint(1, 2000)
            # checks if the item is for sale (id <= 1000)
            if inventory_item_id <= 1000:
                transaction_type = "sale"
                # making sure it hasn't been purchased yet
                if inventory_item_id not in purchased:
                    purchased.append(inventory_item_id)
                    due_date = None
                    if season == 'summer':
                        total_price = random.randrange(150, 350, 5)
                    elif season in ['spring', 'fall']:
                        total_price = random.randrange(100, 300, 5)
                    elif season == 'winter':
                        total_price = random.randrange(50, 200, 5)
                    break
            else:
                transaction_type = "rent"
                # renting can go from 1 hour to 8 hours
                rent_length = random.randint(1,7)
                due_date = datetime_out + timedelta(hours=rent_length)
                
                # inconsistent with hourly price, unnecessary column?
                if season == 'summer':
                    total_price = rent_length * random.randint(30, 50)
                elif season in ['spring', 'fall']:
                    total_price = rent_length * random.randint(20, 40)
                elif season == 'winter':
                    total_price = rent_length * random.randint(10, 30)
                break

        
        element = { 
            'person_id': person_id,     
            'IID': inventory_item_id,
            'type': transaction_type,
            'datetime_out': datetime_out,
            'due_date': due_date,
            'total_price': total_price
            
        }
        elements.append(element)

    return elements
```

# Is component of

```python
# only a very small subset of "other" will be a component of other items
# recall that "other" items have item_id 41..50, the rest have 1..40
def create_is_component_of(num_rows=5):
    elements = []
    
    # making sure no one gets sponsored twice by the same manufacturer
    components = {(k+1): [] for k in range(40, 50)}
    
    for i in range(num_rows):
        
        # logic to prevent the same item to be component twice of the same thing
        while True:
            item_id = random.randint(1, 40)
            component_item_id = random.randint(41, 50)
            if item_id not in components[component_item_id]:
                components[component_item_id].append(item_id)
                break

        
        element = { 
            'item_id': item_id,     
            'component_item_id': component_item_id,

        }
        elements.append(element)

    return elements
```

# Item Color

```python
# color is a multi valued attribute, one item can have many colors.
# we just need to make sure that an item doesn't have the same color several times
# also need to make sure that every item has at least 1 color
# we assign about 3000 colors to the 2000 inventory items
def create_item_color(num_rows=50):
    elements = []
    colors = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'black', 'grey', 'white', 'cyan', 'brown']

    color_assignments = {(k+1): [] for k in range(2000)}
    
    for i in range(num_rows):
        # for the first 2000 rows, assign a color to every item:
        if i < 2000:
            color = random.sample(colors, 1)[0]
            item_id = i+1
            color_assignments[item_id].append(color)
        # after the first 2000, randomly assign more colors 1000 more times
        else:
            # logic to prevent the same item to be component twice of the same thing
            item_id = random.randint(1,2000)
            while True:
                color = random.sample(colors, 1)[0]
                if color not in color_assignments[item_id]:
                    color_assignments[item_id].append(color)
                    break

        
        element = { 
            'item_id': item_id,     
            'color': color,
        }
        elements.append(element)

    return elements
```

```python
# Let's assign about 15 diplomas to the 10 instructors of the DB
# Similar Logic as Color
# Let's assign about 50 plants for the 7 manufacturers. 
def create_teacher_diploma(num_rows=15):
    elements = []
    
    # assigning diplomas to teachers. remember they have PID 1..10
    diplomas = {(k+1): [] for k in range(10)}
    
    for i in range(num_rows):
        # for the first 10 rows, assign a color to every item:
        if i < 10:
            diploma = f"Surf Instructor School of {fake.city()}"
            teacher_id = i+1
            diplomas[teacher_id].append(diploma)
        # after the first 7000, randomly assign more colors 1000 more times
        else:
            # logic to prevent the same item to be component twice of the same thing
            teacher_id = random.randint(1,10)
            while True:
                diploma = f"Surf Instructor School of {fake.city()}"
                if diploma not in diplomas[teacher_id]:
                    diplomas[teacher_id].append(diploma)
                    break

        
        element = { 
            'person_id': teacher_id,     
            'diploma': diploma,
        }
        elements.append(element)

    return elements
```

# Type of Event

```python
# also a multivariate attribute !?
# Similar Logic as Color
# Let's assign about 150 event types to the 100 events
def create_type_of_event(num_rows=150):
    elements = []
    event_types = ['Surf Competition', 'Meet & Greet', 'Food & Drinks', 'Workshop', 
                'Charity', 'Kids Competition', 'Board Customization']
    
    event_assignments = {(k+1): [] for k in range(100)}
    
    for i in range(num_rows):
      
        if i < 100:
            event = random.sample(event_types, 1)[0]
            event_id = i+1
            event_assignments[event_id].append(event)

        else:
            event_id = random.randint(1,100)
            while True:
                event = random.sample(event_types, 1)[0]
                if event not in event_assignments[event_id]:
                    event_assignments[event_id].append(event)
                    break

        
        element = { 
            'event_id': event_id,     
            'type_of_event': event,
        }
        elements.append(element)

    return elements
```
